**Introduction**


The dataset contains information about games that were played by the top 500 tennis players in the world, between 2017 and 2019. The data set, collected by Jeff Sackmann, includes information about the players’ tournament dates, tournament names, minutes, rankings, etc. 


Our first predictor variable is winner aces. An ace in tennis is a serve that “successfully lands in the service box” that the receiver is not able to get their racquet to (Crim, n.d.). The server wins a point. Another predictor variable is the number of successful first serves for winners. On a successful serve, the server gets a point. Our response variable is the tournament level. There are four different tournament levels: G = Grand Slams, M = Masters 1000s, D = Davis Cup, and A = Other tour-level events. 


Using this data, our predictive question is: Can we predict the tournament level using the winner aces and the number of successful first serves for winners on hard surfaces for French players?

**Preliminary Exploratory Data Analysis**


We used two histograms to display our variables of interest from our predictive question with the x-axis for one being Winner Successful Serves and the other being Winner Aces. The y-axis is the frequency of how many times a numerical value was represented for both graphs.


```r
# Loading the Tidyverse Library
library(tidyverse)
```

```r
# Reading the Data from Relative Storage
download <- download.file("https://drive.google.com/u/0/uc?id=12Diw-i1uBQk5a6slIiBbR3F6L14oCYou&export=download", "dataset.csv")
data <- read_csv("dataset.csv")
data
```

```r
# Cleaning the Data

options(repr.plot.width = 12, repr.plot.height = 12)

clean_data <- select(data, tourney_level, winner_ioc, surface, w_ace, w_1stIn, l_ace, l_1stIn) %>%
    filter(surface == "Hard") %>%
    filter(winner_ioc == "FRA") %>%
    filter(!is.na(w_ace))  %>%
    mutate(tourney_level = as_factor(tourney_level))

colnames(clean_data) <- c("tournament_level", "winner_nation", "surface_type", "winner_ace", "winner_firstserve", "loser_ace", "loser_firstserve")

clean_data
```

```r
# Graphing each predictor variable on a Histogram


plot_data_1 <- ggplot(clean_data, aes(x = winner_ace)) + 
    geom_histogram(bins = 30, color = "white", fill = "blue") +
    ggtitle("Frequency of Winner's Aces") +
    labs(x = "Winner's Aces", y = "Frequency") + 
    theme(text = element_text(size = 15))

plot_data_2 <- ggplot(clean_data, aes(x = winner_firstserve)) + 
    geom_histogram(bins = 25,  color = "white", fill = "red") +
    ggtitle("Frequency of Winner's Successful Serves") +
    labs(x = "Winner's Successful Serves", y = "Frequency") + 
    theme(text = element_text(size = 15))



plot_data_1

plot_data_2
```

**Methods**

We will use variables of winner aces and winner successful first serves to predict the tournament level these French players played in. These are our variables of interest because prestigious tournaments such as the Grand Slam and Masters 1000s hosts highly skilled players; since the number of aces and successful serves are indicators of performance levels, using these predictors might result in reliable predictions of the tournament level. Using K-nearest neighbors, we will predict the classification of G, M, D, or A using known values of nearby observations.


To visualize this relationship, we will use a scatter plot to graph winner aces on the x-axis, the successful serves for winners on the y-axis, and differentiate between the different tournament levels using colored points. The surface level plays a role in dictating aces and successful serves, this could act as a confounding variable giving us biased results. Hence, we will choose to examine games with hard surfaces only. 


**Expected Outcomes and Significance**


Using classification prediction, we expect to see that there will be an association between winner aces, the number of successful serves for winners, and the tournament level when playing on hard surfaces. High tournament levels are indicators of higher performance, in regards to the higher number of aces and successful serves. These findings can be used to help professional tennis players. Players can find what level they belong in based on aces they have scored and how many times they have successfully served. Future studies could predict the tournament level by comparing the match score and successful serves. This study begets a question regarding how these predictions vary across different surface types.
